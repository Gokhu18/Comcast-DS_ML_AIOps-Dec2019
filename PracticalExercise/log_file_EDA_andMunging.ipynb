## Join New Data

```python
requests = pd.read_csv('requestall_bucketized_timestamps.csv')
print(requests.shape)
requests.head()
```

```python
nsg_data = pd.read_csv('NSG_Data_bucketized_timestamps.csv')
print(nsg_data.shape)
nsg_data.head()
```

```python
requests.columns
```

```python
templist_ofcolnames = list(nsg_data.columns)
templist_ofcolnames[0] = 'time-rounded'
nsg_data.columns = templist_ofcolnames
nsg_data.columns
```

```python
#df = requests.join(nsg_data,how='left',on='time-rounded', lsuffix='_left', rsuffix='_right')

df = requests.merge(nsg_data,how='left',on='time-rounded')

df.head()
```

```python
df.to_csv('requests_nsg_joined.csv',index=False)
print("done")
```
